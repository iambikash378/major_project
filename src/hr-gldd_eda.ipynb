```python
#Functions for EDA

#Function to explore a number of samples in a grid

def display_grid(datasubset, x, y):

    total_images = x*y
    fig,axs = plt.subplots(2*x,2*y, figsize = (2*y,2*x))
    image_count = 0
        
    match datasubset:
        case "test":
            for i in range(0,2*x,2):
                for j in range(2*y):
                    axs[i,j].imshow(data_trainX[image_count]) 
                    axs[i+1,j].imshow(data_testY[image_count])
                    axs[i,j].axis('off')
                    axs[i+1,j].axis('off')
                    image_count+=1

        case "train":
            for i in range(0,2*x,2):
                for j in range(2*y):
                    axs[i,j].imshow(data_trainX[image_count])
                    axs[i+1,j].imshow(data_trainY[image_count])
                    axs[i,j].axis('off')
                    axs[i+1,j].axis('off')
                    image_count+=1

        case "val":
            for i in range(0,2*x,2):
                for j in range(2*y):
                    axs[i,j].imshow(data_valX[image_count])
                    axs[i+1,j].imshow(data_valY[image_count])
                    axs[i,j].axis('off')
                    axs[i+1,j].axis('off')
                    image_count+=1

```

```python
display_grid('train',2, 3)
```
